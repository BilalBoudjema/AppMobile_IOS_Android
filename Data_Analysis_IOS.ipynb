### Transformation de resultats de l'importation en DataFram

```python
colonnes = ['Nom', 'Nom url','Categorie(s)', 'Date de mise a jour', 'Taille', 'Version Actuelle', 'Version Majeur', 'Version IOS',  'Note', 'Date', 'Heure', 'URL Application', 'Pays', 'Categorie Principale Application','Categorie Secondaire Application']
dataFramTmp = pd.DataFrame( data = values , columns = colonnes) 
dataFramTmp = dataFramTmp.iloc[1: , :]
dataFramTmp.columns
df = dataFramTmp
```

### Affichage de la DataFrame

```python
df.head()
```

* Typage 

```python
df['Nom'] = df['Nom'].astype(object)
df['Nom url'] = df['Nom url'].astype(object)
df['Categorie(s)'] = df['Categorie(s)'].astype(object)
df['Date de mise a jour'] = df['Date de mise a jour'].astype('datetime64[ns]')
df['Taille'] = df['Taille'].astype(object)
df['Version Majeur'] = df['Version Majeur'].astype(float)
df['Version IOS'] = df['Version IOS'].astype(float)
df['Note'] = df['Note'].astype(float)
df['Date'] = df['Date'].astype('datetime64[ns]')
df['Heure'] = df['Heure'].astype('datetime64[ns]')
df['URL Application'] = df['URL Application'].astype(object)
df['Pays'] = df['Pays'].astype(object)
df['Categorie Principale Application'] = df['Categorie Principale Application'].astype(object)
df['Categorie Secondaire Application'] = df['Categorie Secondaire Application'].astype(object)
df.columns
```

* Renommation

```python
category_app = df['Categorie(s)']
nom_app = df['Nom']
note_app = df['Note']
version_ios = df['Version IOS']	
vers_majeur_app = df['Version Majeur']
version_actuelle = df['Version Actuelle']
taille_app = df['Taille']
date_maj_app = df['Date de mise a jour']
pays = df['Pays']
catelogrie_princip = df['Categorie Principale Application']
catelogrie_second = df['Categorie Secondaire Application']
```

### Histogramme : Les version Majeurs en Fonction de Version IOS

```python
fig_ver_android_app = px.histogram( df , x = vers_majeur_app , color = version_ios)
fig_ver_android_app.update_layout(title_text="Les versions majeures en fonction de leurs versions IOS")
fig_ver_android_app.show()
```

### Histogramme : Quantitite d'applications par Notes

```python
fig_app_notes = px.histogram(df, x = df['Note'], template="plotly_dark") 
fig_app_notes.update_layout(title_text="Nombre d'applications iOS par note")
fig_app_notes.show()
```

### Scatter Plot : Les notes des appllications en fonction de leurs versions  

```python
dataFame = px.data.gapminder()
fig = px.scatter(dataFame,
                 x = note_app,
                 y = version_ios,
                 size = note_app,
                 color=note_app,
                 hover_name = version_ios,
                 log_x=True, size_max=18,
                 labels=dict( x = "Note", y ="Version IOS", color = "Degré de notation"),
                 title=" Les notes des applications en fonction de leurs versions iOS",
                 template="plotly_dark")
fig.show()
```

### Voiling Plot : Version IOS Majeurs en fonction de Notes  

```python
fig_voiling = px.violin(df,
                        x= vers_majeur_app ,
                        y = note_app,
                        box=False,
                        points='all',
                        title=" Les versions majeures iOS en fonction de notes",
                        category_orders={'Version_Majeur':[]},
                        color=vers_majeur_app,
                        hover_name=note_app,
                        template="plotly_dark")

fig_voiling.show()
```

### Quantite d'applications telechargees Par Pays

```python
fig_app_pays = px.histogram(df, x = df['Pays'], color = "Pays").update_xaxes(categoryorder="total ascending")
fig_app_pays.update_layout(title_text="Nombre d'application iOS téléchargées par Pays")
fig_app_pays.show()
```

Les categories prinicpales telechargees par les pays

```python
fig_app_pays_cate_princip = px.histogram(df,
                                         x = pays,
                                         color = catelogrie_princip, 
                                         labels=dict( x = "Pays")
                                         ).update_xaxes(categoryorder="total ascending")
fig_app_pays_cate_princip.update_layout(title_text="Les categories principales iOS telechargées par pays")
fig_app_pays_cate_princip.show()
```

Les categories secondaires telechargees par les pays

```python
fig_app_pays_categorie_second = px.histogram(df,
                                             x = pays,
                                             color = catelogrie_second ,
                                             labels=dict( x = "Pays")).update_xaxes(categoryorder="total ascending")
fig_app_pays_categorie_second.update_layout(title_text=" Les categories secondaires telechargées par pays")
fig_app_pays_categorie_second.show()
```
