### Importatation de Google Sheets 

* Depuis url

```python
from __future__ import print_function

import os.path

from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError


from google.oauth2 import service_account

SCOPES = ['https://www.googleapis.com/auth/spreadsheets']
SERVICE_ACCOUNT_FILE = 'Account_Service.json'

creds = None
creds = service_account.Credentials.from_service_account_file(SERVICE_ACCOUNT_FILE, scopes=SCOPES)

# The ID and range of a sample spreadsheet.
SAMPLE_SPREADSHEET_ID = '1gGD4NTP64NB0KkCpbF21WFMT2Vy_6naky41RLF73RKQ'

service = build('sheets','v4', credentials=creds)
# Call the Sheets API
sheet = service.spreadsheets()
result = sheet.values().get(spreadsheetId=SAMPLE_SPREADSHEET_ID,range="Cleaned_Data_Android!A1:Q4410").execute()
values = result.get('values', [])
# print(values)
```

### Transformation de resultats de l'importation en DataFram

```python
colonnes = ['Nom', 'Nom url', 'Developpeur', 'Categorie(s)', 'Nombre installations', 'Date de mise a jour', 'Taille', 'Version Actuelle', 'Version Android', 'Version Majeur',  'Note', 'Date', 'Heure', 'URL Application', 'Pays', 'Categorie Principale Application','Categorie Secondaire Application']
dataFramTmp = pd.DataFrame( data = values , columns = colonnes) 
dataFramTmp = dataFramTmp.iloc[1: , :]
dataFramTmp.columns
df = dataFramTmp
```

### Affichage de la DataFrame

```python
df.head()
```

* Typage 

```python
df['Nom'] = df['Nom'].astype(object)
df['Nom url'] = df['Nom url'].astype(object)
df['Developpeur'] = df['Developpeur'].astype(object)
df['Categorie(s)'] = df['Categorie(s)'].astype(object)
df['Nombre installations'] = df['Nombre installations'].astype(int)
df['Date de mise a jour'] = df['Date de mise a jour'].astype('datetime64[ns]')
df['Taille'] = df['Taille'].astype(object)
df['Version Actuelle'] = df['Version Android'].astype(float)
df['Version Majeur'] = df['Version Majeur'].astype(float)
df['Version Android'] = df['Version Android'].astype(float)
df['Note'] = df['Note'].astype(float)
df['Date'] = df['Date'].astype('datetime64[ns]')
df['Heure'] = df['Heure'].astype('datetime64[ns]')
df['URL Application'] = df['URL Application'].astype(object)
df['Pays'] = df['Pays'].astype(object)
df['Categorie Principale Application'] = df['Categorie Principale Application'].astype(object)
df['Categorie Secondaire Application'] = df['Categorie Secondaire Application'].astype(object)
df.columns
```

* Renommation

```python
category_app = df['Categorie(s)']
nbr_install = df['Nombre installations']
nom_app = df['Nom']
note_app = df['Note']
version_android = df['Version Android']	
vers_majeure_app = df['Version Majeur']
version_actuelle = df['Version Actuelle']
taille_app = df['Taille']
date_maj_app = df['Date de mise a jour']
pays = df['Pays']
catelogrie_princip = df['Categorie Principale Application']
catelogrie_second = df['Categorie Secondaire Application']
```

### Histogramme : Les version Majeurs en Fonction de Version Android

```python
fig_ver_android_app = px.histogram( df , x = vers_majeure_app , color = version_android)
fig_ver_android_app.update_layout(title_text="Les versions majeures en Fonction de leurs version android")
fig_ver_android_app.show()
```

```python
fig_ver_android_app = px.histogram( df , x = vers_majeure_app , color = version_android, template='gridon')
fig_ver_android_app.update_layout(title_text="Les versions majeur en Fonction de leurs version android")
fig_ver_android_app.show()
```

### Scatter Plot : Les notes des appllications en fonction de leurs versions  

```python
dataFame = px.data.gapminder()
fig = px.scatter(dataFame,
                 x = note_app,
                 y = version_android,
                 size = note_app ,
                 color=note_app ,
                 hover_name = version_android,
                 log_x=True, size_max=20,
                 labels=dict( x = "Note", y ="Version Android", color = "Degr de notation "),
                 template="plotly_dark",
                 title=" Les notes des applications Android en fonction de leurs versions android") 
fig.show()
```
